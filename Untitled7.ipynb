```python
import matplotlib.pyplot as plt

# Data points (x, y) - Assuming x is cost and y is probability/likelihood
data_points = {
    142700: 0.56,
    143200: 0.14,
    152000: 0.29
}

# Extract x and y values
x_values = list(data_points.keys())
y_values = list(data_points.values())

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_values, y_values, color='blue', marker='o', label='Risk Points')

# Annotate each point with its y-value
for x, y in data_points.items():
    plt.annotate(f'{y:.2f}', (x, y), textcoords="offset points", xytext=(5,5), ha='left')

# Draw vertical lines connecting each point to the x-axis
for x, y in data_points.items():
    plt.plot([x, x], [0, y], linestyle='--', color='gray', linewidth=0.7)

# Set labels and title
plt.xlabel("Cost/Impact")
plt.ylabel("Probability/Likelihood")
plt.title("Risk Profile with Connections to X-axis")
plt.grid(True, linestyle='--', alpha=0.6)

# Set y-axis limits (adjust as needed based on your data)
plt.ylim(0, max(y_values) + 0.1)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the three scenarios (Probability, ROI at Probability=0 and Probability=1 - approximate from the image)
scenario_data = {
    "Scenario S1": {"probabilities": np.array([0, 1]), "roi": np.array([12.8e6, 13.1e6])},
    "Scenario S2": {"probabilities": np.array([0, 1]), "roi": np.array([10.2e6, 10.4e6])},
    "Scenario S3": {"probabilities": np.array([0, 1]), "roi": np.array([4.9e6, 5.5e6])},
}

# Create the plot
plt.figure(figsize=(10, 6))
plt.title("Sensitivity Analysis", fontsize=14)
plt.xlabel("Probability", fontsize=12)
plt.ylabel("ROI", fontsize=12)
plt.grid(True, linestyle='-', alpha=0.7)

# Plot the ROI for each scenario across the probability range
colors = ['blue', 'orange', 'green']
for i, (scenario, data) in enumerate(scenario_data.items()):
    plt.plot(data["probabilities"], data["roi"], color=colors[i], label=scenario)

# Set y-axis formatting to display in millions
plt.ticklabel_format(style='sci', axis='y', scilimits=(6, 6))
plt.yticks(np.arange(0, 16e6, 2e6)) # Adjust y-ticks to match the image

# Add a legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
